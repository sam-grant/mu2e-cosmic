```python
in_path_natalie = "../../txt/comp/natalie_onspill_original.csv"
df_natalie = pd.read_csv(in_path_natalie)
df_natalie["Absolute Frac. [%]"] = (df_natalie["Events Passing"] / df_natalie["Events Passing"].iloc[0]) * 100
df_natalie["Relative Frac. [%]"] = 100.0
for i in range(1, len(df_natalie)):
    df_natalie.loc[i, "Relative Frac. [%]"] = (df_natalie["Events Passing"].iloc[i] / df_natalie["Events Passing"].iloc[i-1]) * 100
display(df_natalie)
df_natalie.to_csv("../../txt/comp/natalie_onspill.csv", index=False)
```
